## **판다스와 matplotlib**
판다스에서는 .plot()이라는 간단한 메서드로 데이터를 matplotlib으로 시각화하는 방법을 제공한다

```python
import pandas as pd
```

```python
s=pd.Series(np.random.randn(10).cumsum(),index=np.arange(0,100,10))
```

```python
s.plot()
```

데이터 프레임에서도 .plot()메서드가 존재한다. 데이터 프레임의 plot은 여러 컬럼의 값을 동시에 보여줄 수 있다.

```python
df=pd.DataFrame(np.random.randn(10,4).cumsum(0),
                columns=['A','B','C','D'],
                index=np.arange(0,100,10))
```

```python
df.plot()
```

바 차트는 다음과 같이 만들 수 있다.

```python
fig, axes=plt.subplots(2,1)
data=pd.Series(np.random.rand(16),index=list('abcdefghijklmnop'))
data.plot.bar(ax=axes[0],color='k',alpha=0.7)
data.plot.barh(ax=axes[1],color='k',alpha=0.7)
```

데이터 프레임에서 바 차트는 자동으로 그룹화되어 나타난다.

```python
df=pd.DataFrame(np.random.rand(6,4),
                index=['one','two','three','four','five','six'],
                columns=pd.Index(['A','B','C','D'],name='Genus'))
```

```python
df
```

```python
df.plot.bar()
```

다음과 같이 스택 바 차트를 만들 수 있다.

```python
df.plot.barh(stacked=True,alpha=0.5)
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex",col="day",col_wrap=2)
```

# **인터렉티브 시각화**
bokeh와 plotly의 가장 큰 특징은 인터렉티브 시각화를 생성한다는 것이다. matplotlib이 인쇄물에 최적화 되어 있는 반면에 bokeh와 plotly는 브라우저 환경에 특화되어 있다.

## **bokeh**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
```

```python
# 노트북에서 bokeh 플롯을 출력하기 위해 실행한다
output_notebook()
```

```python
N=4000
x=np.random.random(size=N)*100
y=np.random.random(size=N)*100
radii=np.random.random(size=N)*1.5
colors=["#%02x%02x%02x"%(r,g,150) for r, g in zip(np.floor(50+2*x).astype(int),np.floor(30+2*y).astype(int))]

p=figure()
p.circle(x,y,radius=radii,fill_color=colors,fill_alpha=0.6,line_color=None)
show(p)
```

## **Plotly**

```python
import plotly.express as px
df=px.data.iris()
```

```python
fig=px.scatter(df,x="sepal_width",y="sepal_length",color="species")
fig.show()
```

다음과 같이 복잡한 그래프도 그릴 수 있다.

```python
df=px.data.iris()
fig=px.scatter(df,x="sepal_width",y="sepal_length",color="species",marginal_y="violin",
               marginal_x="box",trendline="ols",template="simple_white")
fig.show()
```
